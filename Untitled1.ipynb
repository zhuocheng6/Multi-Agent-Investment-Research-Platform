```python
# def fetch_sec_filings(company_ticker: str) -> str:
#     """Fetches the most recent 10-K filing for a given company ticker."""
#     query = {
#       "query": { "query_string": { "query": f"ticker:{company_ticker} AND formType:\"10-K\"" } },
#       "from": "0", "size": "1", "sort": [{ "filedAt": { "order": "desc" } }]
#     }
#     response = queryApi.get_filings(query)
#     if not response['filings']: return "No 10-K filings found."
#     # In a real app, you would parse the filing's 'linkToFilingDetails'
#     return f"Successfully fetched 10-K filing. The 'Risk Factors' section should be analyzed."
from sec_api import ExtractorApi
import google.generativeai as genai
import os

def fetch_sec_filings(company_ticker: str) -> str:
    """
    Fetches the most recent 10-K filing, extracts the 'Risk Factors' section,
    and summarizes the key risks using an LLM.
    """
    # --- Step 1: Find the latest 10-K filing URL ---
    query = {
      "query": { "query_string": { "query": f"ticker:{company_ticker} AND formType:\"10-K\"" } },
      "from": "0", "size": "1", "sort": [{ "filedAt": { "order": "desc" } }]
    }
    response = queryApi.get_filings(query)
    if not response['filings']:
        return "No 10-K filings found."
    filing_url = response['filings'][0]['linkToFilingDetails']

    # --- Step 2: Extract the 'Risk Factors' section (1A) ---
    extractorApi = ExtractorApi(api_key=os.environ.get("SEC_API_KEY"))
    try:
        risk_factors_text = extractorApi.get_section(filing_url, '1A', 'text')
    except Exception as e:
        return f"Could not extract Risk Factors section: {e}"

    # --- Step 3: Summarize the text with the Gemini API ---
    model = genai.GenerativeModel('gemini-1.5-flash')
    prompt = f"""
    Based on the following 'Risk Factors' text from a company's 10-K filing,
    please identify and summarize the top 3-5 most critical risks.
    Present them as a clear, concise bulleted list.

    TEXT:
    {risk_factors_text}
    """
    try:
        summary = model.generate_content(prompt)
        return summary.text
    except Exception as e:
        return f"Could not summarize the text: {e}"

def fetch_market_news(company_name: str) -> str:
    """Fetches the top 5 recent news headlines for a given company name."""
    articles = newsapi.get_everything(q=company_name, language='en', sort_by='publishedAt', page_size=5)
    if not articles['articles']: return "No news articles found."
    headlines = [article['title'] for article in articles['articles']]
    return "Recent News Headlines:\n- " + "\n- ".join(headlines)

def calculate_volatility(company_ticker: str) -> str:
    """Calculates the 30-day annualized volatility for a given stock ticker."""
    hist = yf.Ticker(company_ticker).history(period="3mo")
    if hist.empty: return "Could not retrieve historical stock data."
    log_returns = np.log(hist['Close'] / hist['Close'].shift(1))
    volatility = log_returns.std() * np.sqrt(252)
    return f"30-Day Annualized Volatility: {volatility:.2%}"

print("✅ Agent tools defined.")
```

```python


AGENT_MODEL = MODEL_GEMINI_2_0_FLASH

sec_agent = Agent(
    name="sec_agent_v1",
    model=AGENT_MODEL,
    description="Provides the latest SEC 10-k filing of the company.",
    instruction="""
You are a financial analyst who ALWAYS follows a specific multi-step process to generate a risk report.
You MUST call all the necessary tools in sequence to gather all the required information before generating the final report.

Your process is as follows:
1.  **SEC Filings:** First, you MUST call the `fetch_sec_filings` tool using the company ticker.
2.  **Market News:** Second, you MUST call the `fetch_market_news` tool using the company name.
3.  **Volatility:** Third, you MUST call the `calculate_volatility` tool using the company ticker.
4.  **Final Report:** After executing ALL THREE tools, you MUST assemble the collected information into a single, final Markdown report.

**Crucially, if any tool fails, you must report the failure in the correct section of the final report but still proceed to the next step.** Do not stop until all tools have been attempted.

Your final output must be a single Markdown report with these exact sections:
- ## Financial Risk Report: [Company Name]
- ### 1. Key Risks from SEC Filings
- ### 2. Market Sentiment from News
- ### 3. Quantitative Volatility
- ### 4. Overall Summary (Synthesize all information you were able to gather)
""",
    # This part was already correct from the last fix
    tools=[fetch_sec_filings, fetch_market_news, calculate_volatility],
)

print(f"Agent '{sec_agent.name}' created using model '{AGENT_MODEL}'.")
```

```python
# @title Setup Session Service and Runner

# --- Session Management ---
# Key Concept: SessionService stores conversation history & state.
# InMemorySessionService is simple, non-persistent storage for this tutorial.
session_service = InMemorySessionService()

# Define constants for identifying the interaction context
APP_NAME = "finance_research_app"
USER_ID = "user_1"
SESSION_ID = "session_001" # Using a fixed ID for simplicity

# Create the specific session where the conversation will happen
session = await session_service.create_session(
    app_name=APP_NAME,
    user_id=USER_ID,
    session_id=SESSION_ID
)
print(f"Session created: App='{APP_NAME}', User='{USER_ID}', Session='{SESSION_ID}'")

# --- Runner ---
# Key Concept: Runner orchestrates the agent execution loop.
runner = Runner(
    agent=sec_agent, # The agent we want to run
    app_name=APP_NAME,   # Associates runs with our app
    session_service=session_service # Uses our session manager
)
print(f"Runner created for agent '{runner.agent.name}'.")
```

```python
# @title Define Agent Interaction Function

from google.genai import types # For creating message Content/Parts

async def call_agent_async(query: str, runner, user_id, session_id):
  """Sends a query to the agent and prints the final response."""
  print(f"\n>>> User Query: {query}")

  # Prepare the user's message in ADK format
  content = types.Content(role='user', parts=[types.Part(text=query)])

  final_response_text = "Agent did not produce a final response." # Default

  # Key Concept: run_async executes the agent logic and yields Events.
  # We iterate through events to find the final answer.
  async for event in runner.run_async(user_id=user_id, session_id=session_id, new_message=content):
      # You can uncomment the line below to see *all* events during execution
      print(f"  [Event] Author: {event.author}, Type: {type(event).__name__}, Final: {event.is_final_response()}, Content: {event.content}")

      # Key Concept: is_final_response() marks the concluding message for the turn.
      if event.is_final_response():
          if event.content and event.content.parts:
             # Assuming text response in the first part
             final_response_text = event.content.parts[0].text
          elif event.actions and event.actions.escalate: # Handle potential errors/escalations
             final_response_text = f"Agent escalated: {event.error_message or 'No specific message.'}"
          # Add more checks here if needed (e.g., specific error codes)
          break # Stop processing events once the final response is found

  print(f"<<< Agent Response: {final_response_text}")
```
